```python
import cv2
from ultralytics import YOLO

# Load model
model = YOLO("/Users/parththapliyal/runs/detect/train2/weights/best.pt")

# Run inference
results = model("Downloads/archive/dataset/images/val/Cars41.png")

# Convert image and display using OpenCV
for result in results:
    img = result.plot()  # Get annotated image
    cv2.imshow("YOLOv8 Detection", img)
    
# Add a waitKey to display the window and allow user interaction
cv2.waitKey(0)  # Press any key to continue
cv2.destroyAllWindows()  # Close the window
```
