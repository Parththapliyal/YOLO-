```python

## Run inference on the new videos
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Open the video file or webcam
video_path = "/Users/parththapliyal/Downloads/3940480-hd_1920_1080_30fps.mp4"  
cap = cv2.VideoCapture(video_path)

# Check if the video opened successfully
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Variables to store the highest confidence frame
highest_confidence = 0
best_frame = None

# Process video frame by frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Stop if the video ends

    # Run YOLOv8 on the frame
    results = model(frame)

    # Process detections
    for result in results:
        annotated_frame = result.plot() 

        # Extract confidence scores from all detections
        if result.boxes is not None:
            for box in result.boxes:
                conf = box.conf[0].item()  # Get confidence score
                x1, y1, x2, y2 = map(int, box.xyxy[0])  

                # Draw confidence score on the frame
                cv2.putText(annotated_frame, f"{conf:.2f}", (x1, y1 - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)

                # Update highest confidence detection
                if conf > highest_confidence:
                    highest_confidence = conf
                    best_frame = annotated_frame.copy()

    # Show the video with detections
    cv2.imshow("YOLOv8 Video Detection", annotated_frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video capture and close OpenCV windows
cap.release()
cv2.destroyAllWindows()

# If a high-confidence frame was found, display it in Jupyter Notebook
if best_frame is not None:
    # Convert BGR (OpenCV format) to RGB (Matplotlib format)
    best_frame_rgb = cv2.cvtColor(best_frame, cv2.COLOR_BGR2RGB)
    
    # Show image in Jupyter Notebook
    plt.figure(figsize=(10, 6))
    plt.imshow(best_frame_rgb)
    plt.axis("off")
    plt.title(f"Highest Confidence Detection: {highest_confidence:.2f}")
    plt.show()
```

```python
# trying to detect number plate using the webcam of the laptop

import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Load the trained YOLOv8 model
model = YOLO("/Users/parththapliyal/runs/detect/train2/weights/best.pt") 

# Open the webcam 
cap = cv2.VideoCapture(0)

# Check if the webcam opened successfully
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

# Process webcam feed frame by frame
while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Failed to capture frame")
        break  # Stop if there's an issue with the webcam

    # Run YOLOv8 on the frame
    results = model(frame)

    # Process detections
    for result in results:
        annotated_frame = result.plot()  # Get annotated frame

        # Extract confidence scores from all detections
        if result.boxes is not None:
            for box in result.boxes:
                conf = box.conf[0].item()  # Get confidence score
                x1, y1, x2, y2 = map(int, box.xyxy[0])  # Get bounding box coords

                # Draw confidence score on the frame
                cv2.putText(annotated_frame, f"{conf:.2f}", (x1, y1 - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)

                # If confidence is high, display the detected number plate in Jupyter Notebook
                if conf > 0.5:  # Adjust confidence threshold if needed
                    plate_crop = frame[y1:y2, x1:x2]  # Crop number plate

                    if plate_crop.size != 0:  # Ensure the cropped image is valid
                        # Convert BGR (OpenCV format) to RGB (Matplotlib format)
                        plate_rgb = cv2.cvtColor(plate_crop, cv2.COLOR_BGR2RGB)

                        # Display the number plate in Jupyter Notebook
                        plt.figure(figsize=(3, 2))
                        plt.imshow(plate_rgb)
                        plt.axis("off")
                        plt.title(f"Detected Plate (Confidence: {conf:.2f})")
                        plt.show()

    # Show the webcam feed with detections
    cv2.imshow("YOLOv8 Real-Time Detection", annotated_frame)

    # Small delay to allow OpenCV to update the window
    cv2.waitKey(1)

    # Press 'q' to exit the webcam feed
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

# Release webcam and close OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
